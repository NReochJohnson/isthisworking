```python
# import 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# load dataset
X = pd.read_csv('faithful.csv')
Y=X.drop(['Unnamed: 0'],axis=1)
df=(Y-Y.mean())/Y.std()
df.head(5)
```

```python
import random
k=2
centroids = pd.DataFrame(np.random.randn(k, 2),columns=['a','b'])
centroids.head(2)
```

```python
m,n=df.shape
df_ind=np.zeros(m)
dist_0=np.zeros(m)
dist_1=np.zeros(m)
for j in range(m):
        dist_0[j]=np.sum((df.values[j,:]-centroids.values[1,:])**2)
        dist_1[j]=np.sum((df.values[j,:]-centroids.values[0,:])**2)
        if dist_0[j] > dist_1[j]:
            df_ind[j]=1
df['ind']=df_ind
df['d_0']=dist_0
df['d_1']=dist_1
            
        
```

```python
plt.scatter(df['eruptions'],df['waiting'],c=df['ind'])
plt.scatter(centroids['a'],centroids['b'],c='red')
plt.show()
```

```python

for i in range(2):
    ah=df[df.ind == i]
    centroids.at[i,'a']=ah['eruptions'].mean()
    centroids.at[i,'b']=ah['waiting'].mean()
print (centroids)
```
